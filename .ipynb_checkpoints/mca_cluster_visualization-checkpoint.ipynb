## MCA Cluster Visualization
This notebook uses a Multiple Correspondence Analysis (MCA) to visually analyze the clustering of job listings. 
The idea is to analyze the scatter plot using the loading values from the first two principal components to determine what skills and technologies are associated with each job role.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mca
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('/my_dataset.csv') 
df = df.reset_index(drop=True)
df.shape
```

```python
df1 = df[df.num_job_labels == 1].copy().sample(frac=1).reset_index(drop=True)
```

```python
'''
Because job listings can contain more than one job role (f.e. one job listing could be hiring for
a Data Scientist, Data Engineer, and Data Analyst), we want to only select the subset of job listings
that have one job role. This is important when doing statistics to measure job trends for certain
job roles.
'''

# Create a new column called job_label and assign an empty string to every row for that column.
df1['job_label'] = ''

# For each row, check which of the job roles was found and assign that job role to the job_label column.
for index, row in df1.iterrows():
    if row['ds_label'] == True:
        df1.at[index, 'job_label'] = 'Data Scientist'
    if row['da_label'] == True:
        df1.at[index, 'job_label'] = 'Data Analyst'
    if row['de_label'] == True:
        df1.at[index, 'job_label'] = 'Data Engineer'
    if row['mle_label'] == True:
        df1.at[index, 'job_label'] = 'Machine Learning Engineer'
    if row['bia_label'] == True:
        df1.at[index, 'job_label'] = 'Business Intelligence Analyst'
```

```python
binary_columns = df1.select_dtypes(include=[float])
binary_columns = binary_columns.drop('work_from_home', axis=1) # lots of nans
binary_columns = binary_columns.drop('max_years_experience', axis=1) # lots of nans
binary_columns = binary_columns.drop('min_years_experience', axis=1) # lots of nans
binary_columns = binary_columns.drop('min_salary', axis=1) # lots of nans
binary_columns = binary_columns.drop('max_salary', axis=1) # lots of nans
int_columns = binary_columns.astype(int)
cluster_df = pd.concat([int_columns, df1[['job_label']]], axis=1)

X = cluster_df.drop('job_label', axis=1)
y = cluster_df['job_label']

print(X.shape, y.shape)
```

```python
y.value_counts()
```

```python
'''
Use chi2 to select top 100 most depended features on the target job roles
'''

selector = SelectKBest(chi2, k=100)

X_new = selector.fit_transform(X, y)

# boolean mask of selected features
mask = selector.get_support()

selected_features = X.columns[mask]

X_new = pd.DataFrame(X_new, columns=selected_features)

print(X_new.shape, y.shape)
```

```python
'''
Because we are working with a large number of binary/categorical variables, we use
Multiple Correspondence Analysis (MCA) instead of Principal Component Analysis.
We select the 2 most important principal components for visualization.
'''

mca_counts = mca.MCA(X_new)
principal_components = mca_counts.fs_r()[:, :2]
```

```python
principal_components.shape
```

```python
# Column_coords contains the loadings of each word/feature for PC1
column_coords = mca_counts.fs_c()

column_coords_df = pd.DataFrame(column_coords, index=X_new.columns)

for index, value in column_coords_df[0].sort_values(ascending=False).iteritems():
    print(f"{index} : {value}")
    
```

```python
# Samething but for PC2
column_coords = mca_counts.fs_c()

column_coords_df = pd.DataFrame(column_coords, index=X_new.columns)

for index, value in column_coords_df[1].sort_values(ascending=False).iteritems():
    print(f"{index} : {value}")
```

```python
'''
This visualization simply plots the job listings base on the two most important principal components.
Each data point/job listing is colored by its true job label
'''

labels = y.tolist()

label_colors = {'Data Scientist': 'tab:cyan', 'Data Analyst': 'tab:olive', 'Data Engineer': 'tab:green', 'Machine Learning Engineer': 'tab:orange', 'Business Intelligence Analyst': 'tab:pink'}

plt.figure(figsize=(16, 12))
plt.scatter(principal_components[:, 0], principal_components[:, 1], c=[label_colors[label] for label in labels], s=3, cmap='viridis', alpha=0.5)

plt.xlabel('\nFirst Principal Component', fontsize=15)
plt.ylabel('\nSecond Principal Component', fontsize=15)
plt.title('MCA Principal Components', fontsize=18)

plt.gca().set_facecolor('#444444')
handles = [plt.Line2D([], [], color=color) for color in label_colors.values()]
labels = list(label_colors.keys())
plt.legend(handles=handles, labels=labels, prop={'size': 10}, loc='upper right')

# plt.savefig("cluster.png", dpi=300)

plt.show()
```
