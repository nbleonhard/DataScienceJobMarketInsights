## Remote Job Proportions Analysis
This notebook answers the follow question, "does there exist an association between job role and the proportion of remote and hybrid jobs?" 
This is achieved via hypothesis testing using the chi squared test of independence with a significance value of 0.05.

A stacked barchart is used to visualize the proportions. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import itertools
import scipy.stats as stats
```

```python
df = pd.read_csv('/my_dataset.csv') 
df = df.reset_index(drop=True)
df.shape
```

```python
'''
Because job listings can contain more than one job role (f.e. one job listing could be hiring for
a Data Scientist, Data Engineer, and Data Analyst), we want to only select the subset of job listings
that have one job role. This is important when doing statistics to measure job trends for certain
job roles.
'''

df1 = df[df.num_job_labels == 1]
df1 = df1.reset_index(drop=True)

# Create a new column called job_label and assign an empty string to every row for that column.
df1['job_label'] = ''

# For each row, check which of the job roles was found and assign that job role to the job_label column.
for index, row in df1.iterrows():
    if row['ds_label'] == True:
        df1.at[index, 'job_label'] = 'Data Scientist'
    if row['da_label'] == True:
        df1.at[index, 'job_label'] = 'Data Analyst'
    if row['de_label'] == True:
        df1.at[index, 'job_label'] = 'Data Engineer'
    if row['mle_label'] == True:
        df1.at[index, 'job_label'] = 'Machine Learning Engineer'
    if row['bia_label'] == True:
        df1.at[index, 'job_label'] = 'Business Intelligence Analyst'
```

```python
job_role_frequencies = df1['job_label'].value_counts()

job_role_frequencies.index = ['Data Analyst', 'Data Scientist', 'BI Analyst', 'Data Engineer', 'ML Engineer']

job_role_frequencies
```

```python
remote_by_job_role = []

job_roles = df1.groupby('job_label')
for job_role in job_roles:
    temp = {}
    
    temp['remote'] = job_role[1][job_role[1].remote == 1.0].shape[0]
    temp['hybrid'] = job_role[1][job_role[1].hybrid == 1.0].shape[0]
    temp['other'] = job_role[1].shape[0] - temp['remote'] - temp['hybrid']
    
    remote_by_job_role.append(pd.Series(temp, name=job_role[0]))
    
```

```python
for role in remote_by_job_role:
    print(role)
```

```python
categories = ['Data Analyst', 'Data Scientist', 'BI Analyst', 'Data Engineer', 'ML Engineer']

remote = []
hybrid = []
other = []
for role in remote_by_job_role:
    remote.append(role['remote'])
    hybrid.append(role['hybrid'])
    other.append(role['other'])


new_order = [1, 3, 0, 2, 4]

# Rearrange the elements
remote = np.array([remote[i] for i in new_order])
hybrid = np.array([hybrid[i] for i in new_order])
other = np.array([other[i] for i in new_order])

# remote = np.array(remote)
# hybrid = np.array(hybrid)
# other = np.array(other)
    
totals = remote + hybrid + other

remote_percentages = remote / totals * 100
hybrid_percentages = hybrid / totals * 100
other_percentages = other / totals * 100

# X-axis values (e.g., different observations)
x = np.arange(len(categories))

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(10, 10))

ax.set_facecolor('#444444')

# Stack bars
ax.bar(x, remote_percentages, label='Remote', color='#b5c2c7')
ax.bar(x, hybrid_percentages, bottom=remote_percentages, label='Hybrid', color='#a6815b')
ax.bar(x, other_percentages, bottom=np.add(remote_percentages, hybrid_percentages), label='Other', color='#4f4f4f')

def add_percentage(ax, cat_percentages, prev_percentages, color):
    for i, percentage in enumerate(cat_percentages):
        ax.text(i, percentage + prev_percentages[i] + 1.8, f"{percentage:.1f}%", ha='center', va='center', color=color, fontsize=19)

add_percentage(ax, remote_percentages, np.zeros_like(remote_percentages), '#b5c2c7')
add_percentage(ax, hybrid_percentages, remote_percentages, '#a6815b')

# Customizing the plot
# ax.set_xlabel('Categories')
ax.set_ylabel('Percentage', fontsize=20, x=1.02)
ax.set_title('Percent of Remote and Hybrid Positions\nBy Job Role', fontsize=20, y=1.02)
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=12)
ax.legend(fontsize=17)

plt.savefig('remote_stacked_bar.png', dpi=300)

# Display the plot
plt.show()

```

```python
contingency_table = {}
for i in range(0, 5):
    contingency_table[remote_by_job_role[i].name] = [value for value in remote_by_job_role[i]]
    
'''
I will use the chi squared test to compare the remote and hybrid proportions across two 
independent job roles. 
H0 : There is no significant association between job role and remote/hybrid proportions.
H1 : There is a significant association between job role and remote/hybrid proportions.
'''

for role1, role2 in list(itertools.combinations(['Business Intelligence Analyst', 'Data Analyst', 'Data Engineer', 'Data Scientist', 'Machine Learning Engineer'], 2)):
    print("{} vs {}".format(role1, role2))
    chi2_stat, p_val, dof, ex = stats.chi2_contingency([contingency_table[role1], contingency_table[role2]])
    alpha = 0.05
    if p_val < alpha:
        print('p-value = {}'.format(p_val))
        print('There is a significant association between job role and remote.\nThe distribution of remote is NOT the same for {} and {}.'.format(role1, role2))
    else:
        print('There is no significant association between job role and remote.\nThe distribution of remote is the same for {} and {}.'.format(role1, role2))
    print('\n')
```

```python
np_remote = np.array([role['remote'] for role in remote_by_job_role])
np_hybrid = np.array([role['hybrid'] for role in remote_by_job_role])
np_other = np.array([role['other'] for role in remote_by_job_role])
not_remote = np_hybrid + np_other

contingency_table = np.array([np_remote, not_remote])

chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

p
```
